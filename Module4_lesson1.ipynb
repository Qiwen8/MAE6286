```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
L = 1.0
nx = 51
dx = L / (nx - 1)
alpha = 1.22e-3
x = numpy.linspace(0.0, L, num = nx)
T0 = numpy.zeros(nx)
T0[0] = 100.0
```

```python
def ftcs(T0, nt, dt ,dx, alpha):
    T = T0.copy()
    sigma = alpha * dt / dx**2
    for n in range(nt):
        T[1:-1] = (T[1:-1] + sigma * (T[2:] - 2.0 * T[1:-1] + T[:-2]))
        T[0] = T0[0]
    return T
```

```python
nt = 1000
sigma = 0.5
dt = sigma * dx**2 / alpha

T = ftcs(T0, nt, dt, dx, alpha)
```

```python
pyplot.figure(figsize=(6.0, 4.0))
pyplot.xlabel('Distance [m]')
pyplot.ylabel('Temperature [C]')
pyplot.grid()
pyplot.plot(x, T, color='C0', linestyle='-', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 100.0);
```

```python
def ftcs2(T0, nt, dt ,dx, alpha, q = 0.0):
    T = T0.copy()
    sigma = alpha * dt / dx**2
    for n in range(nt):
        T[1:-1] = (T[1:-1] + sigma * (T[2:] - 2.0 * T[1:-1] + T[:-2]))
        T[0] = T0[0]
        T[-1] = T[-2] + q * dx
    return T
```

```python
T = ftcs2(T0, nt, dt ,dx, alpha, q = 0.0)
pyplot.figure(figsize=(6.0, 4.0))
pyplot.xlabel('Distance [m]')
pyplot.ylabel('Temperature [C]')
pyplot.grid()
pyplot.plot(x, T, color='C0', linestyle='-', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 100.0);
```
